```python
f = open("C:\\Users\\Seraphina\\Downloads\\202203_202203_연령별인구현황_월간.csv",'r',encoding='cp949')
data = csv.reader(f,delimiter = ',')
header = next(data) #제일 첫번째 element 리스트를 받는다. 

a = '서울특별시'
male=[]
female=[]
for row in data: #각 행을 돌며
    if a in row[0]: #찾는 구에 도착하면
        for i in range(3,104): #거기서 0세부터 100세까지 데이터
            row[i] = int(row[i].replace(',',''))
            male.append((row[i])) #찾는다
        for j in range(-101,0):
            row[j] = int(row[j].replace(',',''))
            female.append((row[j])) #찾는다
        

f.close()

plt.figure(figsize=(5,10))
plt.barh(range(101), female)
plt.barh(range(101), male)
plt.show()
```

# 과제 3

```python
f = open("C:\\Users\\Seraphina\\Downloads\\202203_202203_연령별인구현황_월간.csv",'r',encoding='cp949')
data = csv.reader(f,delimiter = ',')
header = next(data) #제일 첫번째 element 리스트를 받는다. 

a = input("인구 구조를 알고 싶은 광역시나 도의 이름을 입력하세요 (ex: 제주특별자치도):")
male=0
female=0
for row in data: #각 행을 돌며
    if a in row[0]: #찾는 구에 도착하면
        for i in range(3,104): #거기서 0세부터 100세까지 데이터
            row[i] = int(row[i].replace(',',''))
            male+=row[i] #찾는다
        for j in range(106,207):
            row[j] = int(row[j].replace(',',''))
            female+=row[j] #찾는다

f.close()

plt.rc('font', family='Malgun Gothic')    # 그래프에 한글 표시
size = [male,female]          # 데이터
label = ['남','여']       # 레이블
plt.axis('equal')
color = ['crimson','darkcyan']
plt.title(a+'지역의 남녀 성별 비율')
plt.pie(size, labels=label,autopct = '%.1f%%', startangle = 90,colors=color)
plt.show()
```
